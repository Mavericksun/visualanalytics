In mathematics, a [low-discrepancy sequence](https://en.wikipedia.org/wiki/Low-discrepancy_sequence) is a sequence with the property that for all values of N, its subsequence x1, ..., xN has a low [discrepancy](https://en.wikipedia.org/wiki/Equidistributed_sequence#Discrepancy).

Low-discrepancy sequences are also called quasi-random or sub-random sequences, due to their common use as a replacement of uniformly distributed random numbers. The "quasi" modifier is used to denote more clearly that the values of a low-discrepancy sequence are neither random nor pseudorandom, but such sequences share some properties of random variables and in certain applications such as the quasi-Monte Carlo method their lower discrepancy is an important advantage.

We have integrated a kind of low-discrepancy called the [Sobol sequence](https://en.wikipedia.org/wiki/Sobol_sequence), modified from <https://github.com/naught101/sobol_seq>.

# Visualize low-discrepancy sequence and random sequence

We can use plots to show the different characteristics between low-discrepancy sequence and random sequence.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
quasi_rand = sobol(2, 500)
rand = np.random.random((500, 2))
```

```python
plt.plot(quasi_rand[:, 0], quasi_rand[:, 1], 'o', ms=2)
plt.title('Low-discrepancy sequence')
```

```python
plt.plot(rand[:, 0], rand[:, 1], 'o', ms=2)
plt.title('Random sequence')
```

The difference is quite apparent, the low-discrepancy sequence can fill the region more uniformly, thus could be more efficient for monte carlo simulations.
